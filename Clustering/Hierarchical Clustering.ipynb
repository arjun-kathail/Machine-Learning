## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing Dataset

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset.describe()
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
#taken only 2 features for better visualization
X=dataset.iloc[:, [3, 4]].values
```

```python
print(X)
```

## Dendrogram to find optimal number of clusters

```python
from scipy.cluster import hierarchy
dendrogram=hierarchy.dendrogram(hierarchy.linkage(X, method="ward"))
plt.title("Dendrogram")
plt.xlabel("Observation Points")
plt.ylabel("Euclidean Distances")
plt.show()
```

## Training Hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering
clustering=AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage="ward")
y=clustering.fit_predict(X)
y
```

## Visualizations

```python
plt.scatter(X[y==0, 0], X[y==0, 1], s=100, color="red", label="Cluster 1")
plt.scatter(X[y==1, 0], X[y==1, 1], s=100, color="darkblue", label="Cluster 2")
plt.scatter(X[y==2, 0], X[y==2, 1], s=100, color="blue", label="Cluster 3")
plt.scatter(X[y==3, 0], X[y==3, 1], s=100, color="cyan", label="Cluster 4")
plt.scatter(X[y==4, 0], X[y==4, 1], s=100, color="purple", label="Cluster 5")
plt.title("Hierarchical Clustering")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
