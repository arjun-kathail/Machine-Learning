## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing Dataset

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset.describe()
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
#taken only 2 features for better visualization
X=dataset.iloc[:, [3, 4]].values
```

```python
print(X)
```

## Using Elbow method to find the most optimal number of Clusters

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1, 11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(np.arange(1, 11), wcss, color="darkblue")
plt.title("Elbow method")
plt.xlabel("Clusters")
plt.ylabel("WCSS value")
plt.show()
```

## K-Means

```python
kmeans=KMeans(n_clusters=5, init='k-means++', random_state=42)
y=kmeans.fit_predict(X)
y
```

## Visualizations

```python
plt.scatter(X[y==0, 0], X[y==0, 1], s=100, color="red", label="Cluster 1")
plt.scatter(X[y==1, 0], X[y==1, 1], s=100, color="darkblue", label="Cluster 2")
plt.scatter(X[y==2, 0], X[y==2, 1], s=100, color="limegreen", label="Cluster 3")
plt.scatter(X[y==3, 0], X[y==3, 1], s=100, color="cyan", label="Cluster 4")
plt.scatter(X[y==4, 0], X[y==4, 1], s=100, color="purple", label="Cluster 5")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, color="blue", Label="Centroids")
plt.title("K-Means")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
